```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("./ecommerce.csv")
```

```python
data
```

```python
s1 = data['Address'][0]
s2 = data['Address'][1]
```

```python
s2[-14:]
```

```python
data1=data
numeric_colums=data1.select_dtypes(include=[np.number]).columns
numeric_colums
```

```python
correlation_matrix=data1[numeric_colums].corr()
correlation_matrix
```

```python
plt.figure(figsize = (20 , 15))
sns.heatmap(correlation_matrix.corr(),annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
data1.dropna(inplace=True)
```

```python
data1.info()
```

```python
data1=data1.drop(['Email','Address','Avatar'],axis=1)
data1
```

```python
X = data1.drop(['Yearly Amount Spent'], axis=1)
```

```python
y=data1['Yearly Amount Spent']
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
regression_model=LinearRegression()
```

```python
regression_model.fit(X_train,y_train)
```

```python
y_pred = regression_model.predict(X_test)
```

```python
plt.figure(figsize = (15 , 10))
plt.scatter(x=y_test,y=y_test,c='blue',label='Actual',alpha=0.5)
plt.scatter(x=y_test,y=y_pred,c='red',label = 'Predicted',alpha=0.5)
```
